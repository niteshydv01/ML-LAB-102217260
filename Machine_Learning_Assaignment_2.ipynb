```python
from scipy.stats import chi2_contingency, ttest_ind



#T-test for continuous variables
bike_buyers = merged_df[merged_df['BikeBuyer'] == 1]['YearlyIncome']
non_bike_buyers = merged_df[merged_df['BikeBuyer'] == 0]['YearlyIncome']
t_stat, p_val = ttest_ind(bike_buyers, non_bike_buyers)
print(f'T-test p-value for YearlyIncome: {p_val}')

bike_buyers_age = merged_df[merged_df['BikeBuyer'] == 1]['NumberCarsOwned']
non_bike_buyers_age = merged_df[merged_df['BikeBuyer'] == 0]['NumberCarsOwned']
t_stat_age, p_val_age = ttest_ind(bike_buyers_age, non_bike_buyers_age)
print(f'T-test p-value for NumberCarsOwned: {p_val_age}')

# Chi-squared test for categorical variables
contingency_table = pd.crosstab(merged_df['Gender'], merged_df['BikeBuyer'])
chi2, p, dof, ex = chi2_contingency(contingency_table)
print(f'Chi-squared test p-value for Gender: {p}')


# T-test for continuous variables
bike_buyers = merged_df[merged_df['BikeBuyer'] == 1]['AvgMonthSpend']
non_bike_buyers = merged_df[merged_df['BikeBuyer'] == 0]['AvgMonthSpend']
t_stat, p_val = ttest_ind(bike_buyers, non_bike_buyers)
print(f'T-test p-value for AvgMonthSpend: {p_val}')

# Chi-squared test for categorical variables
contingency_table = pd.crosstab(merged_df['City'], merged_df['BikeBuyer'])
chi2, p, dof, ex = chi2_contingency(contingency_table)
print(f'Chi-squared test p-value for City: {p}')


# Chi-squared test for categorical variables
contingency_table = pd.crosstab(merged_df['Education'], merged_df['BikeBuyer'])
chi2, p, dof, ex = chi2_contingency(contingency_table)
print(f'Chi-squared test p-value for Education: {p}')


# Chi-squared test for categorical variables
contingency_table = pd.crosstab(merged_df['NumberChildrenAtHome'], merged_df['BikeBuyer'])
chi2, p, dof, ex = chi2_contingency(contingency_table)
print(f'Chi-squared test p-value for NumberChildrenAtHome: {p}')

# Chi-squared test for categorical variables
contingency_table = pd.crosstab(merged_df['MaritalStatus'], merged_df['BikeBuyer'])
chi2, p, dof, ex = chi2_contingency(contingency_table)
print(f'Chi-squared test p-value for MaritalStatus: {p}')
```

```python
# Determine data type classification
def classify_attribute(df, attribute):
    dtype = df[attribute].dtype
    if pd.api.types.is_numeric_dtype(dtype):
        if len(df[attribute].unique()) > 10:
            # Continuous Data
            return 'Ratio' if df[attribute].min() >= 0 else 'Interval'
        else:
            # Discrete Data
            return 'Ordinal'
    else:
        # Categorical Data
        return 'Nominal'

# Apply classification to selected attributes
for attr in selected_attributes:
    print(f"Attribute: {attr}, Type: {classify_attribute(merged_df, attr)}")
```

##Converting BirthDate into age


```python
from datetime import datetime

# Convert the 'BirthDate' column to datetime format
new_df['BirthDate'] = pd.to_datetime(new_df['BirthDate'], errors='coerce')

# Calculate the age based on today's date
today = datetime.today()
new_df['Age'] = new_df['BirthDate'].apply(lambda x: today.year - x.year - ((today.month, today.day) < (x.month, x.day)))

# Display the updated DataFrame
print(new_df[['BirthDate', 'Age']])
```
