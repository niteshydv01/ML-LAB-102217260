```python
import pandas as pd

# Read the text file
input_file = "/content/retail.txt"

# Prepare a list to store each transaction as a separate list of items
transactions = []

with open(input_file, 'r') as file:
    for line in file:
        # Split each line by whitespace and append as a transaction
        transaction = line.strip().split()
        transactions.append(transaction)

# Convert the list of transactions to a DataFrame
df = pd.DataFrame(transactions)

# Display the first few rows of the DataFrame
print(df.head())
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("/content/transaction.csv", header=None)

# Convert to one-hot encoded format
from mlxtend.preprocessing import TransactionEncoder

transactions = df.values.tolist()
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df_encoded = pd.DataFrame(te_ary, columns=te.columns_)

print("Data prepared for Apriori algorithm.")
```

```python
df.head()
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

# List of minimum support thresholds to test
min_support_counts = [0.1, 0.05, 0.02]  # Example support thresholds (10%, 5%, 2%)

# Find frequent itemsets for each support count
for min_support in min_support_counts:
    print(f"\nFrequent itemsets with minimum support = {min_support}:")

    # Generate frequent itemsets
    frequent_itemsets = apriori(df_encoded, min_support=min_support, use_colnames=True)

    # Filter to only show itemsets with at least two items (optional)
    frequent_itemsets = frequent_itemsets[frequent_itemsets['itemsets'].apply(lambda x: len(x) > 1)]

    # Print the resulting frequent itemsets
    print(frequent_itemsets)

    # Generate association rules if frequent itemsets exist
    if not frequent_itemsets.empty:
        rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
        print("\nAssociation rules:")
        print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])

```
